
Elegimos esta base de datos de una juguetería de modelos a escala porque me parece interesante ver cómo se mueven las ventas en un mercado de coleccionistas. Me sirve para mi carrera porque en el futuro tendré que manejar bases de datos reales y tomar decisiones con números.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Subiendo el archivo de ventas
datos = pd.read_csv('sales_data_sample.csv', encoding='latin1')

# Para ver qué columnas tiene y si hay errores
datos.head()
```

```python
# Primero reviso qué tan sucia está la tabla
print("¿Cuántos datos faltantes hay por columna?")
print(datos.isnull().sum())

print("\n¿Cuántas filas hay antes de limpiar?")
print(len(datos))
```

```python
# 1. Quito las columnas que no voy a usar para que la tabla sea más fácil de leer
tabla_limpia = datos[['SALES', 'QUANTITYORDERED', 'PRICEEACH', 'PRODUCTLINE']].copy()

# 2. Elimino todas las filas que tengan nulos
tabla_limpia = tabla_limpia.dropna()

# 3. Por si acaso, quito si hay algún dato repetido
tabla_limpia = tabla_limpia.drop_duplicates()

print("¡Proceso terminado!")
```

Problema: Detectamos que habían muchos valores nulos en columnas de dirección y contacto que no servían para el cálculo matemático.

```python
# Evidencia de que ya no hay nulos
print("Verificación de nulos después de la limpieza:")
print(tabla_limpia.isnull().sum())

print("\n--- MI TABLA LIMPIA (Primeros 10 registros) ---")
tabla_limpia.head(10)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Creamos el escalador para normalizar los datos entre 0 y 1
scaler = MinMaxScaler()
columnas_num = ['SALES', 'QUANTITYORDERED', 'PRICEEACH']

# Aplicamos la escalación
df_escalado = pd.DataFrame(scaler.fit_transform(tabla_limpia[columnas_num]),
                           columns=columnas_num)

print("✅ Evidencia de Escalación (Datos normalizados entre 0 y 1):")
df_escalado.head()
```

```python
# Calculando las medidas de tendencia central y dispersión
media_v = tabla_limpia['SALES'].mean()
mediana_v = tabla_limpia['SALES'].median()
moda_v = tabla_limpia['SALES'].mode()[0]
varianza_v = tabla_limpia['SALES'].var()
desviacion_v = tabla_limpia['SALES'].std()

print("📊 MI REPORTE ESTADÍSTICO")
print("-" * 30)
print(f"Media (Promedio):    {media_v:.2f}")
print(f"Mediana:             {mediana_v:.2f}")
print(f"Moda:                {moda_v:.2f}")
print("-" * 30)
print(f"Varianza:            {varianza_v:.2f}")
print(f"Desviación Estándar: {desviacion_v:.2f}")
```

```python
# Creamos un resumen organizado de los hallazgos estadísticos
resumen_stats = pd.DataFrame({
    'Medida': ['Media (Promedio)', 'Mediana', 'Moda', 'Varianza', 'Desviación Estándar', 'Rango'],
    'Valor': [
        f"{tabla_limpia['SALES'].mean():.2f}",
        f"{tabla_limpia['SALES'].median():.2f}",
        f"{tabla_limpia['SALES'].mode()[0]:.2f}",
        f"{tabla_limpia['SALES'].var():.2f}",
        f"{tabla_limpia['SALES'].std():.2f}",
        f"{tabla_limpia['SALES'].max() - tabla_limpia['SALES'].min():.2f}"
    ]
})

print("📋 TABLA RESUMEN DE HALLAZGOS")
display(resumen_stats)
```

```python
# Histograma para ver cómo se distribuyen las ventas en el dataset
plt.figure(figsize=(9, 5))
sns.histplot(tabla_limpia['SALES'], kde=True, color='skyblue', edgecolor='black')
plt.title('Distribución de los Montos de Ventas en el Dataset')
plt.xlabel('Monto Facturado')
plt.ylabel('Cantidad de Registros')
plt.show()
```

```python
# Boxplot para identificar visualmente los Outliers detectados en la varianza
plt.figure(figsize=(9, 4))
sns.boxplot(x=tabla_limpia['SALES'], color='lightgreen')
plt.title('Identificación de Outliers (Valores Atípicos)')
plt.xlabel('Rango de Ventas')
plt.show()
```

```python
# Calculando el rango para entender la amplitud de los datos
valor_max = tabla_limpia['SALES'].max()
valor_min = tabla_limpia['SALES'].min()
rango = valor_max - valor_min

print(f"Venta máxima registrada: {valor_max:.2f}")
print(f"Venta mínima registrada: {valor_min:.2f}")
print(f"Rango de los datos: {rango:.2f}")
```

```python
# Gráfico de Violín para ver la densidad de los datos
plt.figure(figsize=(9, 5))
sns.violinplot(x=tabla_limpia['SALES'], color='orchid')
plt.title('Análisis de Densidad de Ventas (Gráfico de Violín)')
plt.xlabel('Monto de Ventas')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

#  PREPARACIÓN: Creamos la "meta".
# Queremos que la IA adivine si una venta es mayor al promedio (1) o menor (0).
promedio = tabla_limpia['SALES'].mean()
tabla_limpia['VENTA_ALTA'] = (tabla_limpia['SALES'] > promedio).astype(int)

# SELECCIÓN: X son las pistas (precio/cantidad) y Y es lo que debe adivinar.
X = tabla_limpia[['PRICEEACH', 'QUANTITYORDERED']]
y = tabla_limpia['VENTA_ALTA']

# DIVISIÓN: 70% para estudiar (train) y 30% para el examen final (test).
# 'random_state' es para que si corro el código de nuevo, la división sea la misma.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# ENTRENAMIENTO: Aquí la IA está "leyendo los libros" del 70% de los datos.
modelo_ia = LogisticRegression()
modelo_ia.fit(X_train, y_train)

# EXAMEN: Le pasamos a la IA los datos que no conoce (X_test) para que adivine.
# Las respuestas de la IA se guardan en 'predicciones'.
predicciones = modelo_ia.predict(X_test)

print("✅ ¡La IA ha terminado su examen!")
```

```python

# COMPARACIÓN: Comparamos la realidad (y_test) contra lo que dijo la IA (predicciones).
matriz = confusion_matrix(y_test, predicciones)

#  DIBUJO: Usamos Seaborn (sns) para que la matriz no sean solo números negros.
plt.figure(figsize=(6,4))
sns.heatmap(matriz, annot=True, fmt='d', cmap='Blues')

# Aquí es donde ponemos los nombres a los ejes para no confundirnos
plt.xlabel('Predicción de la IA (Lo que ella cree)')
plt.ylabel('Realidad del Dataset (Lo que de verdad pasó)')

plt.title('Matriz de Confusión: Evaluación del Modelo')
plt.show()

 # REPORTE: Un resumen de texto con la precisión final.
print("--- REPORTE DE PRECISIÓN ---")
print(classification_report(y_test, predicciones))
```
